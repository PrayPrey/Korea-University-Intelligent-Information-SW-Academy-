```python
df_total_patent_상세103111 = pd.read_csv(direct + 'Total_상세_중간_103111.csv',encoding = 'utf-8-sig',index_col = 0)
```

```python
df_total_patent_상세103111_2 = df_total_patent_상세103111.dropna(subset = ['(21) 출원번호/일자']).reset_index(drop = True).copy()
```

```python
df_total_patent_상세103109_2 = df_total_patent_상세103109.dropna(subset = ['(21) 출원번호/일자']).reset_index(drop = True).copy()
```

```python
df_total_patent_상세_merge = pd.merge(df_total_patent_상세103111_2,df_total_patent_상세103109_2,how = 'outer').copy()
```

```python
df_total_patent_상세_merge.columns
```

```python
df_total_patent_상세_merge_2 = df_total_patent_상세_merge.dropna(subset  =['(21) 출원번호/일자']).drop_duplicates(
    '(21) 출원번호/일자').copy()
```

```python
# ... 중 없는 것 제거
lst_CPC = []
for idx, i in enumerate(df_total_patent_상세_merge_2['(52) CPC '].values):
    w = i.replace('[','').replace("'",'').split(',')
    

    for y in w:
        yy = y.strip().split('/')[0][0:4]
        
        if ('G06Q' in yy) | ('G06N' in yy) | ('H04L' in yy):
            lst_CPC.append(idx)
        
        
```

```python
df_total_patent_상세_merge_2.shape
```

```python
df_total_patent_상세_merge_3 = df_total_patent_상세_merge_2.iloc[list(set(lst_CPC))].copy()
```

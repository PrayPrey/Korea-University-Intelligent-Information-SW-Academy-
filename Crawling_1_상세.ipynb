# 청구항

```python
def make_청구항(i):
    num = i.split(' ')[0]
    url = 'http://kpat.kipris.or.kr/kpat/biblioa.do?method=biblioMain_biblio&next=biblioViewSub04&applno='+str(num)+'&getType=Sub04&link=N'

    res = requests.get(url)

    soup = BeautifulSoup(res.content, 'html.parser')
    
    lst_pre = []
    for i,j in zip(soup.select('td.num'), soup.select('span.wiki_view')):
        lst_pre.append([np.array(i)[0],j.text.replace('\n','')])   
        
    df_pre = pd.DataFrame(lst_pre, columns = ['번호','청구항']).set_index('번호')
    return df_pre.T
```

```python
test_url_2 = 1020210000951

def make_수정(test_url_2):

    url = 'http://kpat.kipris.or.kr/kpat/biblioa.do?method=biblioMain_biblio&next=biblioViewSub01&applno='+str(test_url_2)+'&getType=BASE&link=N'

    res = requests.get(url)

    soup = BeautifulSoup(res.content, 'html.parser')

    lst_all = []

    for i in soup.select('div#divBiblioContent'):

        for j in i.select('div.detial_plan_info ul li'):

            lst_a = []

            if len(j.select('strong')) > 1:
                for jjj in j.select('strong'):
                    lst_a.append([jjj.text,''])

            else:
                if len(j.select('span')) == 1:
                    lst_a.append([j.select_one('strong').text,[jj.text.replace(' ','').replace('\t','').replace('\n','') for jj in j.select('span')]])

                else:
                    lst_b = []
                    for k in j.text.replace('\t','').split('\n'):
                        if len(k.strip()) != 0:
                            lst_b.append(k.strip())
                    lst_a.append(lst_b)


            lst_all.append(lst_a)

        lst_all_2 = [fe[0] for fe in lst_all]


#         lst_all_2.append(['상세정보',i.select_one('span.wiki_view').text.replace('\t','').replace('\n','').strip()])



    lst_all_2[0][1] = [i + ')' for i in lst_all_2[0][1][0].split(')') if len(i) != 0]
    lst_all_2[1][1] = [i + ')' for i in lst_all_2[1][1][0].split(')') if len(i) != 0]
        
        
    dfaa = pd.DataFrame(lst_all_2).set_index(0).T
        
    return dfaa
```
